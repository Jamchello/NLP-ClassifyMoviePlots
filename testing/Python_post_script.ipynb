NB: Mortal Kombat is a martial-arts thriller film. In trainng dataset, the term martial-arts was not hypenated, so decided not to when creating small testing set.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
post_url="http://20.86.144.56/predict"
test_set_file= 'testing_dataset.csv'

ten_movies = pd.read_csv(test_set_file, delimiter=",")

def post_request(movie_plot):
  return [pred.text.strip() for pred in bs(requests.post(post_url, data={'plot': movie_plot}).text, 'html.parser').find_all('li')]

ten_movies["Genre Prediction"] = [" ".join(post_request(movie_plot)) for movie_plot in ten_movies['Plot']]

ten_movies
```
